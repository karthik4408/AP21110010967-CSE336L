```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
def distance(x, y, method):
    if method == 'eu':
        return np.sqrt(np.sum((x - y) ** 2))
    elif method == 'man':
        return np.sum(np.abs(x - y))
    elif method == 'mi':
        return np.max(np.abs(x - y))
    else:
        raise ValueError("Invalid distance method. Choose from 'euclidean', 'manhattan', or 'minkowski'.")
```

```python
x = np.array([10, 5, 3])  
y = np.array([4, 0, 9])  
print(f"Distance using euclidean distance:", distance(x, y, 'eu'))
print(f"Distance using manhattan distance:", distance(x, y, 'man'))
print(f"Distance using minkowski distance:", distance(x, y, 'mi'))
```

```python
iris = load_iris()
X = iris.data  # Features
n_samples = X.shape[0]
distance_matrix = np.zeros((n_samples, n_samples))

for i in range(n_samples):
    for j in range(n_samples):
        distance_matrix[i, j] = distance(X[i], X[j])

print("Euclidean Distance Matrix:")
print(distance_matrix)
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
class_0 = df[df['target'] == 0]
class_1 = df[df['target'] == 1]
```

```python
plt.scatter(class_0['sepal length (cm)'], class_0['sepal width (cm)'], label='Class 0', color='magenta')
plt.scatter(class_1['sepal length (cm)'], class_1['sepal width (cm)'], label='Class 1', color='cyan')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter plot of two classes')
plt.legend()
plt.grid(True)
plt.show()
```
